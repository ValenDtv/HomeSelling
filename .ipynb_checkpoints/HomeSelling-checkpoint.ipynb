```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn import ensemble
from sklearn.naive_bayes import GaussianNB

file_path = 'brooklyn_sales_map.csv'

home_data = pd.read_csv(file_path, nrows=5000)

home_data.shape
```

```python
pd.set_option('display.max_columns', 111)
home_data.describe()
```

```python
home_data.columns.values
```

```python
#Выкидываем не нужные столбцы
home_data.drop([#Дублирующие столбцы
                'building_class', 'building_class_at_sale', 'BldgClass', 'Address', 'borough',
                #Много пустых значений
                'easement', 'apartment_number', 'ZoneDist2', 'ZoneDist3',
                'ZoneDist4', 'Overlay1', 'Overlay2', 'SPDist1', 'SPDist2', 'SPDist3', 'LtdHeight',
                'OwnerType', 'Ext', 'HistDist', 'Landmark', 'ZMCode',
                #Столбцы со многими уникальными значениями и просто не нужные
                'address', 'Borough', 'sale_date', 'Version', 'APPDate', 'Sanborn', 'OwnerName', 'Unnamed: 0',
                'PLUTOMapID', 'FIRM07_FLA', 'PFIRM15_FL', 'MAPPLUTO_F', 'land_sqft', 'gross_sqft'
               ], inplace=True, axis=1)
```

```python
#Оставляем только строки, у которых указан класс
home_data = home_data.dropna(subset=['building_class_category'])
y = home_data['building_class_category']
home_data.drop('building_class_category', inplace=True, axis=1)
#Разделяем признаки на категориальные и количественные
categorical_columns = [c for c in home_data.columns if home_data[c].dtype.name == 'object']
numerical_columns   = [c for c in home_data.columns if home_data[c].dtype.name != 'object']
home_data[categorical_columns].describe()
```

```python
#Заменяем пустые значения на медиану и самые часто встречаемые
home_data = home_data.fillna(home_data.median(axis=0), axis=0)
data_describe = home_data.describe(include=[object])
for c in categorical_columns:
    home_data[c] = home_data[c].fillna(data_describe[c]['top'])
```

```python
#Разделяем категориальные признаки на бинарные и не бинарные
binary_columns    = [c for c in categorical_columns if data_describe[c]['unique'] == 2]
nonbinary_columns = [c for c in categorical_columns if data_describe[c]['unique'] > 2]
#Векторизируем категориальные признаки
for c in binary_columns:
    top = data_describe[c]['top']
    top_items = home_data[c] == top
    home_data.loc[top_items, c] = 0
    home_data.loc[np.logical_not(top_items), c] = 1
data_nonbinary = pd.get_dummies(home_data[nonbinary_columns])
```

```python
#Нормализуем количественные признаки
data_numerical = home_data[numerical_columns]
#data_numerical = (data_numerical - data_numerical.mean()) / data_numerical.std()
scale_features_std = StandardScaler() 
data_numerical = scale_features_std.fit_transform(data_numerical) 

scale_features_mm = MinMaxScaler() 
data_numerical = scale_features_mm.fit_transform(data_numerical)
data_numerical = pd.DataFrame(data_numerical)
```

```python
#Соединям всё в одну таблицу
home_data = pd.concat((data_numerical, home_data[binary_columns], data_nonbinary, y.rename('building_class_category')), axis=1)
#Костыль, что бы привести X и y к одной размерности и убрать все встроки с пустыми значениями, появившиеся после concat
home_data = home_data.dropna()
y = home_data['building_class_category']
home_data.drop('building_class_category', inplace=True, axis=1)
home_data = pd.DataFrame(home_data, dtype=float)
home_data.describe()
```

```python
#Убираем признаки, которые слабо коррелируют с классом
#for c in home_data.columns:
#    if abs(stats.pointbiserialr(home_data[c],y)) < 0.5:
#        home_data.drop((c), inplace=True, axis=1)
#print(home_data.columns)
```

```python
X_train, X_test, y_train, y_test = train_test_split(home_data, y, test_size = 0.3, random_state = 11)
```

```python
#Метод ближайших соседей
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)

err_test  = np.mean(y_test  != knn.predict(X_test))
err_test
```

```python
#Случайный лес
rf = ensemble.RandomForestClassifier()
rf.fit(X_train, y_train)

err_test  = np.mean(y_test  != rf.predict(X_test))
err_test
```

```python
#Градиентный бустинг
gbt = ensemble.GradientBoostingClassifier()
gbt.fit(X_train, y_train)

err_test = np.mean(y_test != gbt.predict(X_test))
err_test
```
